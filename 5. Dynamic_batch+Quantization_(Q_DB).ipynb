```python
import time
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, random_split
from torchvision import datasets, transforms
import matplotlib.pyplot as plt

# --------- Minimal MLP Model ----------
class ShallowMLP(nn.Module):
    def __init__(self):
        super(ShallowMLP, self).__init__()
        self.net = nn.Sequential(
            nn.Flatten(),
            nn.Linear(28 * 28, 64),
            nn.ReLU(),
            nn.Linear(64, 10)
        )

    def forward(self, x):
        return self.net(x)

# --------- Training Utilities ----------
def train(model, loader, optimizer, criterion, epochs=1):
    model.train()
    for _ in range(epochs):
        for x, y in loader:
            optimizer.zero_grad()
            loss = criterion(model(x), y)
            loss.backward()
            optimizer.step()

def evaluate(model, loader):
    model.eval()
    correct, total = 0, 0
    with torch.no_grad():
        for x, y in loader:
            preds = model(x).argmax(dim=1)
            correct += (preds == y).sum().item()
            total += y.size(0)
    return correct / total

# --------- Quantization Utility ----------
def quantize_model_weights(state_dict, num_bits=8):
    q_state = {}
    qmin, qmax = 0, 2**num_bits - 1

    for k, v in state_dict.items():
        v_min, v_max = v.min(), v.max()
        scale = (v_max - v_min) / float(qmax - qmin) if v_max != v_min else 1.0
        zero_point = int(torch.clamp(torch.round(qmin - v_min/scale), qmin, qmax).item())
        q_v = torch.clamp(torch.round(v/scale + zero_point), qmin, qmax).to(torch.uint8)

        q_state[k] = {
            "q_tensor":   q_v,
            "scale":      scale,
            "zero_point": zero_point,
            "orig_shape": v.shape,
        }
    return q_state

# --------- Federated Learning Loop ----------
def federated_training(optimizer_name='adam',
                       max_batch_size=100,
                       num_bits=8,
                       rounds=10,
                       bandwidth_mb_per_s=20):
    num_clients = 10
    client_powers = torch.rand(num_clients)
    max_power = client_powers.max()

    transform = transforms.Compose([
        transforms.ToTensor(),
        transforms.Normalize((0.1307,), (0.3081,))
    ])
    full_dataset = datasets.MNIST(root="./data", train=True, download=True, transform=transform)
    client_data = random_split(full_dataset, [len(full_dataset)//num_clients]*num_clients)

    global_model = ShallowMLP()

    acc_list, comm_list, agg_time_list, sim_time_list, convergence_speed = [], [], [], [], []
    prev_acc = 0

    for rnd in range(1, rounds+1):
        print(f"\n--- Round {rnd} ---")
        q_updates = []

        for i in range(num_clients):
            ratio = (client_powers[i] / max_power).item()
            batch_size_i = max(1, int(ratio * max_batch_size))
            print(f" Client {i}: power={client_powers[i]:.3f} → batch_size={batch_size_i}")

            model = ShallowMLP()
            model.load_state_dict(global_model.state_dict())

            train_loader = DataLoader(client_data[i], batch_size=batch_size_i, shuffle=True)

            if optimizer_name == 'adam':
                optimizer = optim.Adam(model.parameters(), lr=0.001)
            elif optimizer_name == 'sgd':
                optimizer = optim.SGD(model.parameters(), lr=0.001, momentum=0.0)
            elif optimizer_name == 'nag':
                optimizer = optim.SGD(model.parameters(), lr=0.001, momentum=0.9)
            elif optimizer_name == 'mgd':
                optimizer = optim.SGD(model.parameters(), lr=0.001, momentum=0.99)
            else:
                raise ValueError(f"Unknown optimizer: {optimizer_name}")

            train(model, train_loader, optimizer, nn.CrossEntropyLoss(), epochs=1)
            q_dict = quantize_model_weights(model.state_dict(), num_bits=num_bits)
            q_updates.append(q_dict)

        start_comm = time.perf_counter()
        accum = {k: torch.zeros(meta["orig_shape"], dtype=torch.float32)
                 for k, meta in q_updates[0].items()}

        for upd in q_updates:
            for k, meta in upd.items():
                qv = meta["q_tensor"].float()
                deq = (qv - meta["zero_point"]) * meta["scale"]
                accum[k] += deq

        for k in accum:
            accum[k] /= len(q_updates)
        global_model.load_state_dict(accum)
        agg_time = time.perf_counter() - start_comm
        agg_time_list.append(agg_time)

        test_ds = datasets.MNIST(root="./data", train=False, download=True, transform=transform)
        test_loader = DataLoader(test_ds, batch_size=64)
        acc = evaluate(global_model, test_loader)
        acc_list.append(acc)
        print(f" Global Accuracy: {acc:.2f}")

        total_params = sum(p.numel() for p in ShallowMLP().parameters())
        bytes_per_param = num_bits / 8.0
        comm_mb = num_clients * (total_params * bytes_per_param) / 1e6
        comm_list.append(comm_mb)
        sim_time = comm_mb / bandwidth_mb_per_s
        sim_time_list.append(sim_time)
        print(f" Comm: {comm_mb:.2f} MB  |  Net time: {sim_time:.2f}s  |  Agg time: {agg_time:.4f}s")

        if rnd > 1:
            convergence_speed.append(acc - prev_acc)
        prev_acc = acc

    rounds_axis = list(range(1, rounds+1))

    plt.figure(figsize=(8,5))
    plt.plot(rounds_axis, acc_list, 'o-', label='Accuracy')
    plt.plot(rounds_axis, comm_list, 'x--', label='Comm Volume (MB)')
    plt.xlabel('Round'); plt.ylabel('Value')
    plt.title(f'Accuracy & Comm per Round ({optimizer_name.upper()})')
    plt.legend(); plt.grid(True)

    plt.figure(figsize=(8,5))
    plt.plot(rounds_axis[1:], convergence_speed, 's-', label='ΔAccuracy')
    plt.xlabel('Round'); plt.ylabel('Accuracy Change')
    plt.title(f'Convergence Speed ({optimizer_name.upper()})')
    plt.legend(); plt.grid(True)

    plt.figure(figsize=(8,5))
    plt.plot(rounds_axis, agg_time_list, 'd-', label='Agg Time (s)')
    plt.plot(rounds_axis, sim_time_list, '^--', label='Net Time (s)')
    plt.xlabel('Round'); plt.ylabel('Seconds')
    plt.title(f'Time Metrics ({optimizer_name.upper()})')
    plt.legend(); plt.grid(True)

    plt.show()

if __name__ == '__main__':
    for opt in ['adam', 'sgd', 'nag', 'mgd']:
        federated_training(optimizer_name=opt,
                           max_batch_size=100,
                           num_bits=8,
                           rounds=10,
                           bandwidth_mb_per_s=20)
```
