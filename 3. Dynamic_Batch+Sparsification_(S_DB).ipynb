```python
import time
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, random_split
from torchvision import datasets, transforms
import matplotlib.pyplot as plt

# --------- Minimal MLP Model ----------
class ShallowMLP(nn.Module):
    def __init__(self):
        super(ShallowMLP, self).__init__()
        self.net = nn.Sequential(
            nn.Flatten(),
            nn.Linear(28 * 28, 64),
            nn.ReLU(),
            nn.Linear(64, 10)
        )

    def forward(self, x):
        return self.net(x)

# --------- Federated Training Utilities ----------
def train(model, loader, optimizer, criterion, epochs=1):
    model.train()
    for _ in range(epochs):
        for x, y in loader:
            optimizer.zero_grad()
            loss = criterion(model(x), y)
            loss.backward()
            optimizer.step()

def evaluate(model, loader):
    model.eval()
    correct, total = 0, 0
    with torch.no_grad():
        for x, y in loader:
            preds = model(x).argmax(dim=1)
            correct += (preds == y).sum().item()
            total += y.size(0)
    return correct / total

def compress_model_weights(state_dict, compression_ratio=0.1):
    compressed_state = {}
    for k, v in state_dict.items():
        flat = v.view(-1)
        k_val = int(len(flat) * compression_ratio)
        if k_val == 0:
            compressed_state[k] = torch.zeros_like(v)
            continue
        topk_vals, topk_indices = torch.topk(flat.abs(), k_val)
        mask = torch.zeros_like(flat)
        mask[topk_indices] = 1
        compressed_flat = flat * mask
        compressed_state[k] = compressed_flat.view(v.shape)
    return compressed_state

# --------- Main Federated Learning Process ----------
def federated_training(optimizer_name='adam',
                       clients_per_round=None,
                       max_batch_size=100,
                       compression_ratio=0.1,
                       rounds=10,
                       bandwidth_mb_per_s=20):
    num_clients = 10
    client_powers = torch.rand(num_clients)
    max_power = client_powers.max()

    transform = transforms.Compose([
        transforms.ToTensor(),
        transforms.Normalize((0.1307,), (0.3081,))
    ])
    full_dataset = datasets.MNIST(root="./data", train=True, download=True, transform=transform)
    client_data = random_split(full_dataset, [len(full_dataset) // num_clients] * num_clients)

    global_model = ShallowMLP()

    acc_list = []
    comm_list = []
    comm_time_list = []
    simulated_time_list = []
    convergence_speed = []
    prev_acc = 0

    for rnd in range(1, rounds + 1):
        print(f"\n--- Round {rnd} ---")
        local_models = []

        for i in range(num_clients):
            ratio = (client_powers[i] / max_power).item()
            batch_size_i = max(1, int(ratio * max_batch_size))
            print(f" Client {i}: power={client_powers[i]:.3f} → batch_size={batch_size_i}")

            model = ShallowMLP()
            model.load_state_dict(global_model.state_dict())

            train_loader = DataLoader(client_data[i], batch_size=batch_size_i, shuffle=True)

            if optimizer_name == 'adam':
                optimizer = optim.Adam(model.parameters(), lr=0.001)
            elif optimizer_name == 'sgd':
                optimizer = optim.SGD(model.parameters(), lr=0.001, momentum=0.0)
            elif optimizer_name == 'nag':
                optimizer = optim.SGD(model.parameters(), lr=0.001, momentum=0.9)
            elif optimizer_name == 'mgd':
                optimizer = optim.SGD(model.parameters(), lr=0.001, momentum=0.99)
            else:
                raise ValueError(f"Unknown optimizer: {optimizer_name}")

            train(model, train_loader, optimizer, nn.CrossEntropyLoss(), epochs=1)

            compressed = compress_model_weights(model.state_dict(), compression_ratio)
            local_models.append(compressed)

        start_comm = time.perf_counter()
        new_state = {}
        for key in global_model.state_dict().keys():
            stack = torch.stack([m[key] for m in local_models])
            new_state[key] = torch.mean(stack, dim=0)
        global_model.load_state_dict(new_state)
        agg_time = time.perf_counter() - start_comm
        comm_time_list.append(agg_time)

        test_ds = datasets.MNIST(root="./data", train=False, download=True, transform=transform)
        test_loader = DataLoader(test_ds, batch_size=64)
        acc = evaluate(global_model, test_loader)
        acc_list.append(acc)
        print(f" Global Accuracy: {acc:.2%}")

        total_params = sum(p.numel() for p in ShallowMLP().parameters())
        comm_mb = num_clients * (total_params * compression_ratio * 4) / 1e6
        comm_list.append(comm_mb)
        sim_time = comm_mb / bandwidth_mb_per_s
        simulated_time_list.append(sim_time)
        print(f" Comm: {comm_mb:.2f} MB  |  simulated net time: {sim_time:.2f}s  |  agg time: {agg_time:.4f}s")

        if rnd > 1:
            convergence_speed.append(acc - prev_acc)
        prev_acc = acc

    rounds_axis = list(range(1, rounds + 1))

    plt.figure(figsize=(8, 5))
    plt.plot(rounds_axis, acc_list, marker='o', label=f'{optimizer_name.upper()} Accuracy')
    plt.plot(rounds_axis, comm_list, marker='x', label='Comm Volume (MB)')
    plt.xlabel('Round')
    plt.ylabel('Value')
    plt.title(f'Accuracy & Comm per Round ({optimizer_name.upper()})')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()

    plt.figure(figsize=(8, 5))
    plt.plot(rounds_axis[1:], convergence_speed, marker='s', label='ΔAccuracy')
    plt.xlabel('Round')
    plt.ylabel('Accuracy Change')
    plt.title(f'Convergence Speed ({optimizer_name.upper()})')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()

    plt.figure(figsize=(8, 5))
    plt.plot(rounds_axis, comm_time_list, marker='d', label='Agg Time (s)')
    plt.plot(rounds_axis, simulated_time_list, marker='^', label='Net Time (s)')
    plt.xlabel('Round')
    plt.ylabel('Seconds')
    plt.title(f'Time Metrics ({optimizer_name.upper()})')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()

    plt.show()

if __name__ == '__main__':
    for opt in ['adam', 'sgd', 'nag', 'mgd']:
        federated_training(optimizer_name=opt,
                           max_batch_size=100,
                           compression_ratio=0.1,
                           rounds=10,
                           bandwidth_mb_per_s=20)
```
